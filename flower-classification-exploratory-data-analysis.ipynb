# Flower Classification Exploratory Data Analysis

## Importing Libraries

```python
import os
from glob import glob
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from PIL import Image
```

Accesssing the files in the dataset, the dataset contain images.

```python
images = glob("/kaggle/input/flower-classification-dataset/files/*.jpg")
images[:5]
```

```python
df = pd.read_csv("/kaggle/input/flower-classification-dataset/labels.csv")
df.head()
```

```python
df.tail()
```

```python
count = df.category.value_counts() #counts all the value in the category and places the highest first
count
```

```python
plt.figure(figsize = (20, 10))
df.category.value_counts().plot(kind = "bar") #Same as above now plot into a bar graph
```

Printing the highest and lowest classes and their values

```python
print(f"The Highest class {count.index[0]} contains {count.values[0]} images")
print(f"The Lowest class {count.index[-1]} contains {count.values[-1]} images")
```

Categorizing unique elements in the dataset

```python
unique = df.iloc[df.category.drop_duplicates().index].sort_values('category').reset_index(drop = True)
unique
```

Mapping images to the classes

```python
fig, axs = plt.subplots(34,3, figsize = (15, 225))
ctr = -1
for i, j in enumerate(unique.values):
    if i%3==0:
        ctr+=1
    axs[ctr, i%3].imshow(Image.open(f"/kaggle/input/flower-classification-dataset/files/{j[0]}.jpg").resize((200,200)))
    axs[ctr, i%3].set_title(str(j[1]))
```

Visualizing individual class item in dataset

```python
target_class = 40
```

```python
fig, axs = plt.subplots(15,3, figsize=(15, 75))
ctr =-1
for i, j in enumerate(df[df.category == target_class].head(45).values):
    if i%3==0:
        ctr+=1
    axs[ctr, i%3].imshow(Image.open(f"/kaggle/input/flower-classification-dataset/files/{j[0]}.jpg").resize((200,200)))
    axs[ctr, i%3].set_title(str(j[1]))
```
